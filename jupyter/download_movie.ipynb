## 步骤
1. m3u8文件
2. 将m3u8文件的内容转换为ts文件的url
3. 将ts片段拼接起来 -> 构成完整的视频
4. 保存到本地

```python
import requests
import re
```

```python
video_name = '我想吃掉你的胰脏.mp4'
```

```python
m3u8_url = 'https://v6.dious.cc/20220527/TC9DnlzS/1500kb/hls/index.m3u8'
resp = requests.get(m3u8_url)
resp_text = resp.text
print(resp_text)
```

```python
url_postfixs = re.findall('/.*',resp_text)
print(url_postfixs)
```

https://v6.dious.cc/20220527/TC9DnlzS/1500kb/hls/2gD7i51L.ts

```python
url_prefix = 'https://v6.dious.cc'
```

```python
url_ts_list = [ url_prefix + url_postfix for url_postfix in url_postfixs ]
```

```python
url_ts_list[0]
```

```python
open(video_name,'w').close()
with open(video_name,'ab') as file:
    for url_ts in url_ts_list:
        res_bytes = requests.get(url_ts,timeout=60).content
        file.write(res_bytes)
        file.flush()
```

```python
filename = 'output.mp4'
open(filename,'w').close()
url_ts_list_1 = url_ts_list[1:]
with open(filename,'ab') as file:
    for url_ts in url_ts_list:
        res_bytes = requests.get(url_ts,timeout=60).content
        file.write(res_bytes)
        file.flush()
```
